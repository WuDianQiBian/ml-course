## 2\. 数值与基本运算

## 3\. 类型与类型转换

```python
x = 3
y = 3.0
z = "3"
```

```python
print(type(x), type(y), type(z))  # 查看类型
```

```python
print(isinstance(x, int), isinstance(y, int))  # isinstance 用于判断类型；type(...) 用于拿到类型
```

```python
# 常见类型转换:

print(x, "->", str(x))     # int -> str
print(y, "->", int(y))     # float -> int
print(z, "->", float(z))   # str -> float
```

```python
# Python 中变量是弱类型的，可以在运行时重新绑定为不同类型

x = 3
print(x, type(x))

x = 3.5
print(x, type(x))

x = "three"
print(x, type(x))
```

## 6\. 布尔运算

```python
a = True
b = False

print(a and b)       # 与：只有两边都为 True，结果才为 True
print(a or b)        # 或：只要有一边为 True，结果就是 True
print(a ^ b)         # 异或：两边不相同才为 True
print(not a, not b)  # 非
```

```python
x = 0
y = []
z = None

print(bool(x), bool(y), bool(z))  # 0 / 空容器 / None 通常为 False
```

```python
x = 1
y = [1, 2, 3]
z = "abc"

print(bool(x), bool(y), bool(z))  # 非0数字 / 非空容器 / 非空对象 通常为 True
```

```python
print(x or 999)  # 常见默认值写法：若 x 为 falsy，则取后面的值 999
```

```python
print(z is None)  # 判断 None 推荐用 is None
```

```python
print(3 != 4, 3 == 3)  # 不等 / 相等判断
```

```python
print(3 > 2, 3 >= 3, 3 < 4, 3 <= 2)  # 大小比较
```

```python
print(2 < 3 < 4)  # 连续比较
```

## 7\. 条件判断：if / elif / else

```python
score = 85

if score >= 90:
    level = "A"
elif score >= 80:
    level = "B"
else:
    level = "C"

print(level)
```

## 8\. list 的增删改查

```python
lst = [10, 20, 30, 40, 50]
```

```python
print(lst[0], lst[-1])  # 访问：首元素与末元素
```

```python
print(lst[0:3])    # 切片：左闭右开
```

```python
print(lst[0:5:2])  # 间隔切片：[start, stop, step]
```

```python
lst.append(60)  # 追加 1 个元素
print(lst)
```

```python
lst.extend([70, 80])  # 追加多个元素
print(lst)
```

```python
lst.insert(1, 15)  # 在指定位置插入
print(lst)
```

```python
lst[2] = 999  # 修改元素
print(lst)
```

```python
popped = lst.pop()  # 删除并返回末元素
print(popped)
print(lst)
```

```python
lst.remove(15)  # 按值删除（只删第一个匹配）
del lst[0]  # 按下标删除
print(lst)
```

```python
print(lst.index(30), lst.count(20))  # 查找位置 / 计数
```

## 9\. 循环 与 enumerate / zip

```python
for i in range(5):  # range 生成 0..4
    print(i)
```

```python
i = 0
while i < 5:  # while 循环生成0..4
    print(i)
    i += 1
```

```python
names = ["a", "b", "c"]
for name in names:  # for 循环遍历列表
    print(name)
```

```python
for i, name in enumerate(names):  # enumerate 提供 (index, value)
    print(i, name)
```

```python
values = [10, 20, 30]
for name, v in zip(names, values):  # zip 并行遍历多个序列
    print(name, v)
```

## 10\. 列表推导式（List comprehension）

```python
xs = [1, 2, 3, 4, 5]
```

```python
squares = [x * x for x in xs]
print(squares)
```

```python
mapped = [f"item_{x}" for x in xs]
print(mapped)
```

```python
evens = [x for x in xs if x % 2 == 0]
print(evens)
```

```python
points = [(x, y) for x in range(3) for y in range(3)]
print(points)
```
